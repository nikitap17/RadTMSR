```python
# Lemmatization

import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('punkt')
nltk.download('wordnet')

# Function to lemmatize a token
def lemmatize_token(token):
    lemmatizer = WordNetLemmatizer()
    lemmatized_token = lemmatizer.lemmatize(token)
    return(lemmatized_token)
```

**Text pre-processing pipeline:**
`Docs` --> `replacement_Docs` --> `cleanDocs`-->`final_Docs`


```python
# Define a dictionary of replacements
replacements = {
    'isation': 'ization',
    # 'polarisation': 'polarization',
    # 'socialisation': 'socialization',
    # 'mobilisation': 'socialisation',
    # 'stigmatisation': 'stigmatization',
    # 'securitisation': 'securitization',
    'behaviour':'behavior',
    'islamic state': 'isis',
    'far-right': 'far right',
    'far-left': 'far left',
    'countering violent extremism':'cve'
}
```

```python
# Function to perform replacements based on the dictionary and replace country names

def replace_keywords(doc):
    new_doc = []
    for kw in doc:
        if kw in country_names:
            # kw = 'region'  # Directly replace country names with 'REGION'
            kw = ''  # Remove all country names
        else:
            for old, new in replacements.items():
                kw = kw.replace(old, new)
        new_doc.append(kw)
    return new_doc

# Apply the replacement function to each document
replacement_Docs = [replace_keywords(doc) for doc in Docs]

# Output the updated list of documents
replacement_Docs
```

```python
# Lemmatize documents
cleanDocs = []
for doc in replacement_Docs:
  clean_document = []
  for word in doc:
    if (word not in exclusion) and (word not in stoplist):
      tokens = word.split(" ")
      clean_document.append(" ".join([lemmatize_token(word) for word in tokens]))
  cleanDocs.append(clean_document)



cleanDocs
```

```python
# Punctuation replacements
punc_rep = {
    '-': '&&',
    'â€“ ': '&&',
    ' ': '_',
    '/': '',
    "'s":'',
    "s'":'',
    "shi'a":'shia',
    '.':'',
}

final_Docs = []

for article in cleanDocs:
    new_article = []
    for keyword in article:
        for old, new in punc_rep.items():
            keyword = keyword.replace(old, new)
        new_article.append(keyword)
    final_Docs.append(new_article)

print(final_Docs)  # Properly display the final result
```

```python
import pandas as pd
import re

# Function to normalize keywords by replacing spaces, hyphens, and specific characters with underscores
def normalize_keyword(keyword):
    # Replace spaces and hyphens with underscores
    # keyword = keyword.replace(" ", "_").replace("-", "_")
    # Replace "(", ")", "<", and ">" with underscores using regular expression
    keyword = re.sub(r"[()\/<>]", "_", keyword)
    return keyword

# Step 1: Normalize the keywords in each document
normalized_keywords = [[normalize_keyword(kw) for kw in doc] for doc in final_Docs]
# Step 2: Convert the list of lists into a pandas Series
keywords_series = pd.Series(normalized_keywords)
# Step 3: Explode the Series to create a single list of keywords and count occurrences
keyword_vocabulary = keywords_series.explode().value_counts()

# Export to excel
# pd.DataFrame(keyword_vocabulary).to_excel('/content/drive/MyDrive/2024/1__ In progress/MyRTMR/1__Kws__wordlist.xlsx')
pd.DataFrame(keyword_vocabulary).to_excel('1__Kws__wordlist.xlsx')
```

#### Pruning: Modelling of word frequency with negative binomial distribution

```python
# Model a negative binomial distribution based on term frequency

import numpy as np
from scipy.stats import nbinom

def nbinom_percentile(freq,percentile):
  # Estimate the parameters using the method of moments
  mean, var = np.mean(freq), np.var(freq)
  p = 1 - (var - mean) / var
  r = mean**2 / (var - mean)
  neg_binom = nbinom(r, p)
  # Calculate the target percentile percentile of the fitted distribution
  target_percentile = neg_binom.ppf(percentile)

  return(target_percentile)
```

```python
nbinom_percentile(keyword_vocabulary.values,0.95)
```

#### Topic modeling using latent Dirichlet allocation for subcorpora detection

```python
from gensim.corpora import Dictionary

# Convert documents to a list of tokenized words
docs_tokenized = cleanDocs
# Create a dictionary representation of the documents
dictionary = Dictionary(docs_tokenized)
# Filter out words
dictionary.filter_extremes(keep_n = len(selected_keywords))
# Create a bag-of-words representation of the documents
corpus = [dictionary.doc2bow(doc) for doc in docs_tokenized]
# Increase the number of passes and iterations
passes = 10  # Increase the number of passes
iterations = 50  # Increase the number of iterations
```

```python
from gensim.models import LdaModel, CoherenceModel
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import time

# Define the range of number of topics
min_topics = 1
max_topics = 50

# Initialize lists to store evaluation metrics
perplexity_scores = []
coherence_scores = []

# Start running time
start = time.time()
# Iterate over the range of number of topics
for num_topics in range(min_topics, max_topics + 1):
    loop_start = time.time()

    # Run LDA
    lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=num_topics, passes=passes, iterations=iterations, random_state=42)
    # Evaluate perplexity
    perplexity = lda_model.log_perplexity(corpus)
    perplexity_scores.append(perplexity)
    # Evaluate coherence score
    coherence_model = CoherenceModel(model=lda_model, texts=docs_tokenized, dictionary=dictionary, coherence='c_v')
    coherence = coherence_model.get_coherence()
    coherence_scores.append(coherence)

    loop_end = time.time()
    print('N. topics: {}'.format(num_topics))
    print('Coherence score: {}'.format(round(coherence,4)))
    print('Running time: {}[s]'.format(round(loop_end-loop_start,2)))

# End running time
end = time.time()
print('Running time: {}[s]'.format(round(end-start,2)))

# Create subplots
from plotly.subplots import make_subplots
fig = make_subplots(rows=1, cols=2, subplot_titles=('Perplexity vs. Number of Topics', 'Coherence Score vs. Number of Topics'))
# Adding Perplexity plot
fig.add_trace(go.Scatter(x=list(range(min_topics, max_topics + 1)), y=perplexity_scores, mode='lines+markers', name='Perplexity'),row=1, col=1)
# Adding Coherence Score plot
fig.add_trace(go.Scatter(x=list(range(min_topics, max_topics + 1)), y=coherence_scores, mode='lines+markers', name='Coherence Score'),row=1, col=2)
# Update xaxis properties
fig.update_xaxes(title_text='Number of Topics', row=1, col=1)
fig.update_xaxes(title_text='Number of Topics', row=1, col=2)
# Update yaxis properties
fig.update_yaxes(title_text='Perplexity', row=1, col=1)
fig.update_yaxes(title_text='Coherence Score', row=1, col=2)
# Update layout and show plot
fig.update_layout(height=500, width=1000, title_text="Evaluation Metrics by Number of Topics", showlegend=False)
fig.show()
```
